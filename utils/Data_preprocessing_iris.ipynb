Libraries

```python
import sklearn as sk
from sklearn import datasets,model_selection
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
```

Dataset

```python
# Download the iris dataset from a public URL
import requests

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
response = requests.get(url)

# Save the dataset to a local file
with open("iris.csv", "w") as f:
    f.write(response.text)

# Load the dataset into a pandas DataFrame
iris = pd.read_csv('iris.csv', names=["sepal length", "sepal width", "petal length", "petal width", "class"])
iris # Our example dataset
```

```python
# Split Train test dataset
train,test = sk.model_selection.train_test_split(iris,test_size=0.2,shuffle=True)
```

Cleaning

```python
# Displaying the shape of the train and test dataframes
print("Shape of train dataset:", train.shape)
print("Shape of test dataset:", test.shape)

# Checking for missing values
print("\nMissing values in train dataset:\n", train.isnull().sum())
print("\nMissing values in test dataset:\n", test.isnull().sum())

# Checking data types
print("\nData types of train dataset:\n", train.dtypes)
print("\nData types of test dataset:\n", train.dtypes)

# Displaying descriptive statistics
print("\nDescriptive statistics of train dataset:\n", train.describe())
print("\nDescriptive statistics of test dataset:\n", test.describe())

# Dropping duplicate rows (if any) - Iris dataset has no duplicates, this is for demonstration
train_cleaned = train.drop_duplicates()
test_cleaned = test.drop_duplicates()
print("\nShape of train dataset after dropping duplicates:", train_cleaned.shape)
print("Shape of test dataset after dropping duplicates:", test_cleaned.shape)

# Example of renaming columns (if needed)
train_renamed = train.rename(columns={'sepal length (cm)': 'sepal_length'})
print("\nTrain dataset with renamed column:\n", train_renamed.head())

# Example of handling outliers (e.g., using IQR - not strictly necessary for this dataset)
numerical_train = train.select_dtypes(include=np.number) # Select only numerical columns for quantile calculation

Q1 = numerical_train.quantile(0.25)
Q3 = numerical_train.quantile(0.75)
IQR = Q3 - Q1
train_outliers_removed = train[~((numerical_train < (Q1 - 1.5 * IQR)) |(numerical_train > (Q3 + 1.5 * IQR))).any(axis=1)]
print("\nShape of train dataset after removing potential outliers:", train_outliers_removed.shape)
```
